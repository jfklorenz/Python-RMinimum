RMinimum : Phase 4 - Jupyter

```python
# Imports
import queue
import math
```

Testfall : $W = [0, \cdots, n-1]$, $n0$

```python
# User input
n = 2**10
n0 = 2**32
case = 0
# Case = {0: user input, 1: |W'| < log2(n0)^2, 2:|W'| = log2(n0)^2, 3: |W'| > log2(n0)^2} 

# Automatic
if case == 1:
    n = 2**10
    n0 = 2**32 + 1
elif case == 2:
    n = 2**10
    n0 = 2**32
elif case == 3:
    n = 2**10
    n0 = 2**32 - 1

Wnew = [i for i in range(n)]
cnt = [0 for _ in range(n)]

# Show Testcase
print('          Testcase:          ')
print('=============================')
print("W' = [0, ..., " + str(n - 1) + ']')
print('log2(n0)   =', round(math.log(n0, 2), 4))
print('log2(n0)^2 =', round(math.log(n0, 2)**2, 4))
print('=============================')
```

Algorithmus : Phase 4

```python
def phase4(Wnew, n0, cnt):

    if len(Wnew) <= math.log(n0, 2)**2:
        q = queue.Queue()

        for ele in Wnew:
            q.put(ele)
        while q.qsize() > 1:
            a = q.get()
            b = q.get()

            if a < b:
                q.put(a)
            else:
                q.put(b)

            cnt[a] += 1
            cnt[b] += 1
        mini = q.get()
        
        return mini, cnt
    
    else:
        return False, cnt

# ==================================================
# Testfall
mini, cnt = phase4(Wnew, n0, cnt)
```

Resultat :

```python
def test(Wnew, n0, mini, cnt):
    print('')
    print('Testfall n / n0:', len(Wnew), '/', n0)
    print('====================================')
    print('Recursive call:')
    print('---------------')
    print("|W'|       :", len(Wnew))
    print('log2(n0)^2 :', math.log(n0, 2)**2)
    print("|W'| <= log2(n0)^2 :", len(Wnew) <= math.log(n0, 2)**2)
    print('====================================')
    print('Fragile Complexity:')
    print('-------------------')
    print('min(f_w(n)) :', min([cnt[w] for w in Wnew]))
    print('max(f_w(n)) :', max([cnt[w] for w in Wnew]))
    print("log2(|W'|)  :", round(math.log(len(Wnew), 2), 2))
    print('====================================')
    print('Process:')
    print('--------')
    print('Minimum :', mini)
    print('====================================')
    return

# ==================================================
# Testfall
test(Wnew, n0, mini, cnt)
```
